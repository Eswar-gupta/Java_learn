### Basics of Java
#### Java is a "Write Once, Run Anywhere" (WORA) Language
- **This is how Java runs**

- **Java is a "Write Once, Run Anywhere" (WORA) Language**
    - **Binary Portability:** Unlike C++, where you must generate a separate executable (`.exe`) for every OS, Java compiles to **Bytecode** (`.class` or `.jar`).
    - **One File for All:** You can share the exact same `.jar` file to any computer (Windows, Mac, Linux). The JVM acts as a universal adapter that translates this file for the specific hardware.
    - If you send a .exe (c++ excutable generated on linxu to windows it won't work

- **Comparison of Execution Flows:**
    - Source Code (.cpp) $\xrightarrow{\text{Compiler}}$ Machine Code (.exe / .out) $\xrightarrow{\text{Direct}}$ CPU
    - Source Code (.py) $\xrightarrow{\text{Interpreter}}$ Internal Bytecode $\xrightarrow{\text{Line-by-Line}}$ CPU
    - Source Code (.java) $\xrightarrow{\text{Compiler (javac)}}$ Bytecode (.class) $\xrightarrow{\text{JVM + JIT}}$ Machine Code $\rightarrow$ CPU
    
                      ---------
                      |  JVM  |  <-- The Universal Adapter
                      ---------
                          ↓
                      ---------
                      |   OS  |
                      ---------
                          ↓
                      ---------
                      |Hardware|
                      ---------

### Classes and objects Introduction
#### Defining Classes & File Management (The "100 Files" Scenario)

**Context:** Addressed the scenario of a project with a `Main` class and 100 dependency classes (like `Bank`, `User`, `Database`, etc.) to understand how each language manages definitions, compilation, and linking.

##### 1. How C++ Handles It (The "Manual Builder")
* **Split Files:** Code is physically split into **Headers (`.h`)** (Declarations/Menu) and **Source (`.cpp`)** (Implementation/Recipe). You manage pairs of files.
* **Inclusion (`#include`):** * This is a literal **copy-paste** operation. The preprocessor takes the text from `Bank.h` and pastes it into `Main.cpp`. 
    * *Crucial Rule:* You never include `.cpp` files, or you will get "Multiple Definition Errors."
* **Compilation (The List):** * You must explicitly list **every single source file** to the compiler (e.g., `g++ Main.cpp Bank.cpp User.cpp ...`). 
    * If you miss one file, the linker will fail.
* **Static Linking:** * The Linker stitches all 100 compiled object files (`.o`) together *before* the program runs.
    * **Result:** One giant, standalone executable file (e.g., `myApp.exe`) containing all the code.

##### 2. How Java Handles It (The "Smart Librarian")
* **Single File:** No headers. A class is defined entirely in one `.java` file.
* **Inclusion (`import`):** * `import com.bank.Bank;` does **not** copy code. It acts as a directory lookup, telling the compiler "The class `Bank` is located in the package `com.bank`."
* **Smart Compilation:** * You usually only compile the entry point: `javac Main.java`.
    * The compiler is smart enough to see that `Main` uses `Bank`, find `Bank.java` on the disk, and compile it automatically.
* **Dynamic Loading (No "Linker"):** * Java does not create one giant file. It creates 101 separate small `.class` files.
    * The **JVM** loads these files from the hard drive *dynamically* while the program is running, only when they are actually needed.

#### Executables & Portability

**Why Java is "Write Once, Run Anywhere":**

* **C++ (Hardware-Dependent - The "Custom Furniture"):**
    * **The Output:** Creates **Machine Code** (0s and 1s) specific to the OS and CPU (e.g., Windows/Intel) where it was compiled.
    * **Transfer:** You cannot send this `.exe` to a Linux or Mac user; it won't run. 
    * **Solution:** You must send them the **Source Code**, and they must re-compile it to build a new executable that fits their specific machine.

* **Java (JVM-Dependent - The "IKEA Instructions"):**
    * **The Output:** Creates **Bytecode** (`.class` files). This is a universal "intermediate" language, not machine code.
    * **Transfer:** You send the `.class` files to anyone (Windows, Mac, Linux). You do *not* send source code.
    * **Execution:** The receiver's **Java Virtual Machine (JVM)** acts as a translator. It reads the universal Bytecode and translates it into native machine code for *that specific computer* on the fly.

#### Strings and importances of keyword "new" and stack vs heap
**stack**
- It is the temporary memory in c++ all things like int,flot,double etc...
- If you do (In c++) double* d = {3.12,2.14,5.0}; --> it stored on stack and scope of d dies after the function and we can't return d to anthor fuction

**Heap**
- The perment memory place of c++ things don't get destoryed and new key word needed (c++)double* d = new double[3]{3.12,2.14,5.0}; --> Lies in Heap so can be returned when ever it is needed (return d;)

**In Java**
- Category A: new is NEVER used
    - Primitives: int, float, double, boolean, char, long, short, byte.
    - These are the only things that stay on the Stack as raw values.

- Category B: new is "Hidden" (Optional to type, but it happens)
    - For these special classes, Java provides "Literals" or "Factory Methods" so you don't have to type new, even though an object is created on the Heap:
    - Strings: String name = "Gemini"; (Hidden new via the String Pool).
    - Arrays: int[] d = {1, 2, 3}; (Hidden new for the array object).
    - Wrapper Classes: Integer x = 10; (Hidden new via Autoboxing).

- When your OS runs any program, it creates a **process** — an isolated bubble with its own:
    - Memory (RAM allocated to it)
    - Its own Stack, Heap, Metaspace (in Java's case)
    - A process ID (PID)

#### Same Package Subclass , Non-subclass ,, Different Package Subclass , Non-subclass
**Same Package Subclass (ChildSame.java)**

This class lives in the same folder and inherits from Parent.
Access: It can see everything except private.
Example: If Parent has a protected or default variable, ChildSame can use it directly as if it were its own.

**Same Package Non-subclass (Neighbor.java)**

This class is in the same folder but does not inherit from Parent. It just creates an instance (object) of Parent.
Access: It can see public, protected, and default.
Example: Inside Neighbor, you write Parent p = new Parent(); p.variable;. This works for everything except private.

**Different Package Subclass (ChildDiff.java)**

This class is in a different folder (pack2) but inherits from Parent.
Access: It can see public and protected. It cannot see default (package-private).
Example: It can use protected variables from Parent via inheritance (using super or just the variable name), but it cannot see "default" variables because it's in a different "neighborhood."

**Different Package Non-subclass (Stranger.java)**

This is the most restricted. It is in a different folder and has no family relation to Parent.
Access: It can only see public members.
Example: Stranger can only use things marked public. protected, default, and private are all invisible to it.
